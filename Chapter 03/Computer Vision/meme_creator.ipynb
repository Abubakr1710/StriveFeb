```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('images/minion.jpg')
type(img)
```

```python
rgb_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cimg= rgb_img.copy()
plt.figure(figsize=(20,15))
cv2.putText(cimg,'When we have extra day off', (130,15), cv2.FONT_HERSHEY_COMPLEX,0.5, (0,255,0))
plt.imshow(cimg)
plt.savefig('meme.png')
```
