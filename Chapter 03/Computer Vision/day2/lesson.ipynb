```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('img/dog.jpg')
type(img)
```

```python
#in BGR format
plt.figure(figsize=(12,8))
plt.imshow(img)
plt.title('BGR Format')
plt.show()
```

```python
# in RGB format
rgb_img=img.copy()
rgb_img=cv2.cvtColor(rgb_img,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12,8))
plt.imshow(rgb_img)
plt.title('RGB Format')
plt.show()
```
